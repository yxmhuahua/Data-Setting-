```python
import os
import xml.etree.ElementTree as ET
import numpy as np
import matplotlib.pyplot as plt
#import cv2
def XmlFindId(xmladdr, id):
    bbox = []
    tree = ET.parse(xmladdr)
    i = 0
    for obj in tree.findall('object'):
        if obj.find('name').text != str(id):
            continue
        else:
            bndbox_anno = obj.find('bndbox')
            bbox = [int(bndbox_anno.find(tag).text) for tag in ('xmin', 'ymin', 'xmax', 'ymax')]
            i = i + 1
    if i > 1:
        print("warning! duplicate ID in same image",xmladdr)

    return bbox


def checkduplicateID(root):
    root, _, files = next(os.walk(root))
    for i in range(-1,504):
        for file in files:
            addr = os.path.join(root,file)
            XmlFindId(addr,i)
        print(i," done")
#    for i in range(1,6):

def showBboxInID(id):
    idw=id
    bboxlist = []
    imglist = []
    namelist=[]
    i=1
    print(i)
    root, _, files = next(os.walk(r"xml20190815_c"+str(i)+"_s3"))
    for file in files:
        addr = os.path.join(root, file)

        bbox = XmlFindId(addr, id)
        if bbox != []:
            bboxlist.append(bbox)
            img = plt.imread(r"20190815_c"+ str(i)+ "_s3/" + file.split('.')[0] + r".jpg")
            imglist.append(img[bbox[1]:bbox[3],bbox[0]:bbox[2],:])
            namelist.append(file[0:4])
    for img,name in zip(imglist,namelist):
        plt.imshow(img)
        plt.axis('off')
        imname=str(id)
        
        #imname=imname.zfill(4)
        
        imname=os.path.join(str(idw).zfill(4)+r'_c'+str(i)+r's3_0'+name+".jpg")
        print(imname)
        plt.savefig(r'./query_box1/'+imname)
        #cv2.imwrite(img,r'./query_box/'+imname)
```
